```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
time = np.arange(0, 10, 0.1)
sin_amp = np.sin(time)
cos_amp = np.cos(time)
```

```python
plt.plot(time, sin_amp, label='sine')
plt.plot(time, cos_amp, label='cosine')

plt.legend()
plt.axhline(color='black')
plt.show()
```
